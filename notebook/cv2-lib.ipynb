```python
from realsense_recorder.camera.utils import list_camera_configurations
import platform
```

```python
current_os = platform.system()
print(f"Detected OS: {current_os}")
print("Scanning for connected cameras...\n")

cameras = list_camera_configurations()

if not cameras:
    print("No video capture devices found.")
else:
    for cam in cameras:
        print(f"=== Camera Index {cam['index']} ===")
        print(f"Backend Engine: {cam['method']}")

        caps = cam["capabilities"]

        if cam["method"] == "V4L2 Hardware Query":
            for pixel_fmt, resolutions in caps.items():
                print(f"\n  Pixel Format: [{pixel_fmt}]")
                for res, fps_list in resolutions.items():
                    fps_str = ", ".join(f"{fps} FPS" for fps in sorted(fps_list))
                    print(f"    - Resolution: {res:<12} -> Supported FPS: [{fps_str}]")
        else:
            print("\n  Supported Modes:")
            for item in caps:
                fps_str = (
                    ", ".join(f"{fps} FPS" for fps in item["framerates"])
                    if item["framerates"]
                    else "Default Driver Rate"
                )
                print(f"    - Resolution: {item['resolution']:<12} -> Supported FPS: [{fps_str}]")

        print("\n" + "=" * 50)
```
